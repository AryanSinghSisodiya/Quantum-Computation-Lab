```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0,1)
qc.ccx(0,1,2)
qc.measure_all()
qc.draw("mpl")
```

```python
import numpy as np
theta = np.pi/4
qc1 = QuantumCircuit(3)
qc1.x(0)
qc1.h(1)
qc1.z(2)
qc1.crx(theta,0,1)
qc1.mcry(theta,[0,1],2)
qc1.draw("mpl")
```

```python
qc1.width() #No. of qubits
```

```python
qc1.depth()
```

```python
from qiskit.quantum_info import Statevector
qc2 = QuantumCircuit(2)
qc2.x(0)
qc2.x(1)
state = Statevector.from_instruction(qc2)
print(state)
```

```python
from qiskit.visualization import plot_state_city ,plot_histogram
plot_state_city(state)
```

```python
state2 = state.evolve(qc2)
print(state2)
```

```python
qc3 = QuantumCircuit(4)
qc3.x(0)
qc3.z(1)
qc3.cx(1,2)
qc3.crz(np.pi/12,0,3)
qc3.rx(-np.pi/7,2)
qc3.draw("mpl")
```

```python
state1 = Statevector.from_instruction(qc3)
print(state1)
```

```python
plot_state_city(state1)
```

```python
prob = state1.probabilities_dict(None,2)
plot_histogram(prob)
```
